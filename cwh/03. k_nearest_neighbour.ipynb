```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = datasets.load_iris()

# Assigning features and labels from datasets
features = iris.data
labels = iris.target
```
