# Train a logistic regression classifier to predict whether a flower is <i>Iris virginica <i> or not

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
# Load dataset
iris = datasets.load_iris()

# Assigning features and labels from datasets
X = iris['data'][:, 3:]
Y = (iris['target'] == 2).astype(int) # astype used because classifier doesn't understand true or false
```

```python
# Train a logistic regression classifier
clf = LogisticRegression()
clf.fit(X,Y)
```

```python
#example
clf.predict(([[1.6]]))
```

```python
#example 2
clf.predict(([[2.6]]))
```

```python
X_new = np.linspace(0,3,1000).reshape(-1,1)
print(X_new)
```

```python
Y_prob = clf.predict_proba(X_new)
```

```python
plt.plot(X_new,Y_prob[:,1],"g-", label="virginica")
plt.show()
```
